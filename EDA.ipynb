IMPORTING THE PACKAGES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset we extract

```python
df = pd.read_csv('C:/SQL/Full_Track.csv')
```

BASIC INFO

```python
print("Basic Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

Check for missing values

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Top 5 popular tracks
```

```python
top_tracks = df.sort_values(by='Popularity', ascending=False).head(5)
print("\nTop 5 Most Popular Tracks:")
print(top_tracks[['Track Name', 'Artist', 'Popularity']])
```

Most frequent artists

```python
top_artists = df['Artist'].value_counts().head(5)
print("\nTop 5 Most Frequent Artists:")
print(top_artists)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(data=top_tracks, x='Popularity', y='Track Name', palette='viridis')
plt.title('Top Tracks by Popularity')
plt.xlabel('Popularity')
plt.ylabel('Track Name')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Popularity'], bins=10, kde=True, color='skyblue')
plt.title('Popularity Distribution')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Duration (minutes)', color='lightgreen')
plt.title('Track Duration Distribution')
plt.xlabel('Duration (minutes)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Artist', order=df['Artist'].value_counts().head(5).index, palette='Set2')
plt.title('Top 5 Artists by Number of Tracks')
plt.xlabel('Track Count')
plt.ylabel('Artist')
plt.show()
```

```python
import seaborn as sns

# Correlation matrix
corr = df[['Popularity', 'Duration (minutes)']].corr()

# Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.savefig('correlation_matrix.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_tracks, x='Popularity', y='Track Name', palette='viridis')
plt.title('Top Tracks by Popularity')
plt.xlabel('Popularity')
plt.ylabel('Track Name')
plt.tight_layout()
plt.savefig('top_tracks_by_popularity.png')  # Saves the plot
plt.show()
```
